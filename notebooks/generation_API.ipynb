```
from svg_schematic import Schematic, Resistor, Capacitor, Inductor, Source, Wire
from inform import Error, error, os_error


with Schematic(filename='circuit_test.svg'):
    for node in netlist:
        components = netlist[node]["components"]
        
        nodes = []
        for component in components:

            componentKey = next(iter(component))
            location = component[componentKey]["location"]
            orientation = component[componentKey]["orientation"]

            # Reistor
            if componentKey == "resistor":
                symbol = Resistor(C=location, orient=orientation)
            
            # Inductor
            elif componentKey == "inductor":
                symbol = Inductor(C=location, orient=orientation)
                
            # Capacitor
            elif componentKey == "capacitor":
                symbol = Capacitor(C=location, orient=orientation)

            # Power Supply
            elif componentKey == "source":
                symbol = Source(C=location, orient=orientation, kind="vdc")

            # Diode
            elif componentKey == "diode":
                symbol = Resistor(C=location, orient=orientation)

            # Component not recognized
            else:
                continue

            # getting negative or positive end of component
            if node == component[componentKey]["nodes"][0]:
                nodes.append(symbol.p)
            elif node == component[componentKey]["nodes"][1]:
                nodes.append(symbol.n)
            else:
                continue

            Wire(nodes, "-|-")

                
```
